# Hypothetizing questions

This process is lead by a quest to formulate questions, rather than find possible answers. In order to allow for questions to emerge from data, we use the process we term "hypothesizing" which describes an observation of importance of certain terms through the measure of their appearance in scientific writing (something we consider documented in goodle n-gram project)

```python
import datetime
import pandas as pd
# import pandas.io.data
import numpy as np
from matplotlib import pyplot as plt
import sys
import getngrams as ng 
# from pandas import Series, DataFrame
# import xkcd as fun
%matplotlib inline
```

## N-grams for the initial list of keywords
This list of keywords is used to gather tweets in the keyword-based collection process. It is also used to identify profiles to stream. Some of these words do not appear in google n-gram data (DynamoBIM, Rhinoceros 3D, Rhino 3D, Grasshopper 3D). 

We look at the books published since 1960 (there is little mention of the terms from the list prior to 1960 and their meaning might have changed too)

We show two groups of words: software and hardware tools, which help identify discourse of practitioners:

```python
keywords = ['BIM','CAD', 'AutoCAD', 'Revit', 'Autodesk', 'Internet of Things', 'smart home']
A = pd.DataFrame()
for i in range(len(keywords)):
    try: 
        tmp = ng.runQuery('-nosave -noprint -startYear=1960 -smoothing=3 -endYear=2008 -caseInsensitive '+keywords[i])
#         A['year']=tmp.year.values[:]
        weights = tmp.values[:,1:]
        mx  = np.max(weights,axis=0)
        mn  = np.min(weights,axis=0)
        R = mx-mn
        weights = (weights-mn)/R
        tmp.loc[:,1:]=weights
        A[keywords[i]]=tmp.values[:,1]

    except:
        print(keywords[i], 'not enough data')
        
fig = plt.figure()
for i in range(len(keywords)):
    try: 
        plt.plot(tmp.year,A[keywords[i]],linewidth=1,label=keywords[i])
        xticks = np.arange(tmp.year[0],2011,10).astype(int)
        plt.xticks(xticks)
#         plt.yticks([])
    except:
        print(keywords[i], 'not enough data')

# 
plt.legend(loc='best',bbox_to_anchor = (1.0, 1.0),fontsize = 'large')
fig.set_size_inches(14,7)
```

## Some Fun

```python
import numpy as np
import matplotlib.pyplot as plt
N = 400
x1= np.random.normal(loc=17,scale=5,size=N)[:,np.newaxis]
x2= np.random.normal(loc=0,scale=5,size=N)[:,np.newaxis]
y = 3*x1 + np.random.normal(loc=.0, scale=.4, size=N)[:,np.newaxis]

y =-.1*x1**3 + 2*x1*x1 + 2*np.sqrt(x1)+ 10*np.random.normal(loc=30.0, scale=4.7, size=len(x1))[:,np.newaxis]

def polynomial_regr(degree=1):
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn import linear_model
   
    X_tr = x1[:].astype(float)
    y_tr = y[:].astype(float)
    
    poly = PolynomialFeatures(degree=degree)
    X_tr_ = poly.fit_transform(X_tr)
    
    regr = linear_model.LinearRegression()
    regr.fit(X_tr_, y_tr)
    
    
    y_pred_tr = regr.predict(X_tr_)[:]
    
    plt.plot(X_tr,y_tr,'.r',markersize=10,alpha=.4 );
    plt.plot(X_tr,y_pred_tr,'.b',markersize=10,alpha=.4 );
from ipywidgets import interact, HTML, FloatSlider
interact(polynomial_regr,degree=(1,50,1));
```
